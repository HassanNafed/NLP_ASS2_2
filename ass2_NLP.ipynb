```python
from transformers import pipeline

# Initialize the text generation pipeline with the desired model
generator = pipeline('text-generation', model='gpt2')

# Generate document 1 for machine learning
prompt1 = "machine learning"
generated_text1 = generator(prompt1, max_length=70)

# Generate document 2 for artificial intelligence
prompt2 = "artificial intelligence"
generated_text2 = generator(prompt2, max_length=70)

# Generate document 3 for natural language processing
prompt3 = "natural language processing"
generated_text3 = generator(prompt3, max_length=70)

# Generate document 4 for computer vision
prompt4 = "computer vision"
generated_text4 = generator(prompt4, max_length=70)

# Print or save the generated documents
print("Document 1 (Machine Learning):", generated_text1)
print("Document 2 (Artificial Intelligence):", generated_text2)
print("Document 3 (Natural Language Processing):", generated_text3)
print("Document 4 (Computer Vision):", generated_text4)
```

```python

from bs4 import BeautifulSoup
import re
import nltk
from nltk.stem import WordNetLemmatizer
nltk.download('punkt')  # Download the 'punkt' package for tokenization
nltk.download('wordnet')  # Download WordNet for lemmatization
nltk.download('stopwords')  # Download stop words
from nltk.corpus import stopwords
# Initialize WordNet Lemmatizer and stopwords list
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def process_text(text):
    # Step 1: Clean data by removing symbols and characters
    cleaned_text = re.sub(r'[^\w\s]', '', text)

    # Step 2: Normalize text to lower case
    normalized_text = cleaned_text.lower()

    # Step 3: Tokenize the text
    tokens = nltk.word_tokenize(normalized_text)

    # Step 4: Lemmatization
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]

    # Step 5: Remove stop words
    filtered_tokens = [word for word in lemmatized_tokens if word not in stop_words]

    # Get unique words
    unique_words = set(filtered_tokens)

    return {
        'cleaned_text': cleaned_text,
        'normalized_text': normalized_text,
        'tokens': tokens,
        'lemmatized_tokens': lemmatized_tokens,
        'filtered_tokens': filtered_tokens,
        'unique_words': unique_words
    }

# Process generated documents
processed_documents = {}
all_processed_texts = []
for i, (prompt, generated_text) in enumerate(zip([prompt1, prompt2, prompt3, prompt4], [generated_text1, generated_text2, generated_text3, generated_text4]), 1):
    processed_text = process_text(generated_text[0]['generated_text'])
    processed_documents[f'Document {i}'] = processed_text
    all_processed_texts.append(processed_text)
# Print processed documents
for doc, processed_doc in processed_documents.items():
    print(f"{doc}:")
    # print("Cleaned Text:", processed_doc['cleaned_text'])
    # print("Normalized Text:", processed_doc['normalized_text'])
    # print("Tokens:", processed_doc['tokens'])
    # print("Lemmatized Tokens:", processed_doc['lemmatized_tokens'])
    # print("Filtered Tokens:", processed_doc['filtered_tokens'])
    print("Unique Words:", processed_doc['unique_words'])
    print("\n")
```

```python
#tf_idf from scratch
import pandas as pd

import math
from collections import Counter

# Step 1: Get TF for each word for all documents
def compute_tf(documents):
    tf_values = []
    for doc in documents:
        tf_dict = {}
        counter = Counter(doc['unique_words'])
        total_words = len(processed_documents)
        for word, count in counter.items():
            tf_dict[word] = count / total_words
        tf_values.append(tf_dict)
    return tf_values
print(tf_values)
# Step 2: Get IDF for each word
def compute_idf(documents):
    idf_dict = {}
    total_docs = len(documents)
    for doc in documents:
        for word in doc['unique_words']:
            idf_dict[word] = idf_dict.get(word, 0) + 1
    for word , val in idf_dict.items():
        idf_dict[word] = math.log(total_docs+1 / float(val)+1)+1
    return idf_dict

# Step 3: Multiply TF * IDF
def compute_tfidf(tf_values, idf_values):
    tfidf_values = []
    for tf_dict in tf_values:
        tfidf_dict = {}
        for word, tf in tf_dict.items():
            tfidf_dict[word] = tf * idf_values[word]
        tfidf_values.append(tfidf_dict)
    return tfidf_values

# Step 4: Compute sum of squares of all TF-IDF values for each document
def compute_sum_of_squares(tfidf_values):
    sum_of_squares = []
    for tfidf_dict in tfidf_values:
        squares_sum = sum(val ** 2 for val in tfidf_dict.values())
        sum_of_squares.append(squares_sum)
    return sum_of_squares

# Step 5: Get Normalized TFIDF
def normalize_tfidf(tfidf_values, sum_of_squares):
    normalized_tfidf_values = []
    for tfidf_dict, sum_squares in zip(tfidf_values, sum_of_squares):
        normalized_tfidf_dict = {}
        for word, tfidf in tfidf_dict.items():
            normalized_tfidf_dict[word] = tfidf / math.sqrt(sum_squares)
        normalized_tfidf_values.append(normalized_tfidf_dict)
    return normalized_tfidf_values

# Example usage:
# Compute TF, IDF, and TF-IDF
tf_values = compute_tf(all_processed_texts)
idf_values = compute_idf(all_processed_texts)
tfidf_values = compute_tfidf(tf_values, idf_values)

# Compute sum of squares of all TF-IDF values
sum_of_squares = compute_sum_of_squares(tfidf_values)

# Compute normalized TF-IDF
normalized_tfidf_values = normalize_tfidf(tfidf_values, sum_of_squares)

# Print the normalized TF-IDF values
for i, tfidf in enumerate(normalized_tfidf_values, 1):
    print(f"Document {i} Normalized TF-IDF:")
    print(tfidf)
    print("\n")
tfidf_df = pd.DataFrame(normalized_tfidf_values)

# Print TF-IDF values
print(tfidf_df)
```

```python
#tf_idf built in
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# Assuming the processed_texts is a list of strings (documents)
processed_texts = [' '.join(list(doc['unique_words'])) for doc in all_processed_texts]

# doc['unique_words']
# Initialize the TfidfVectorizer
vectorizer = TfidfVectorizer()

# Fit and transform the processed texts
tfidf_matrix = vectorizer.fit_transform(processed_texts)

# Get feature names
feature_names = vectorizer.get_feature_names_out()
```
